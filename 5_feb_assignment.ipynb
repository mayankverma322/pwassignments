```python
# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
# A class is a blueprint or a template that defines the attributes and behaviors of objects. An object is an instance of a class, which represents a real-world entity.
# Example:

class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def drive(self):
        print(f"Driving {self.brand} {self.model}")


my_car = Car("Toyota", "Camry")
my_car.drive()


# Q2. Name the four pillars of OOPs.
# The four pillars of Object-Oriented Programming are:
# 1. Encapsulation: Encapsulating data and methods into a single unit (class) to hide complexity and protect data.
# 2. Inheritance: Creating new classes (derived classes) from existing classes (base classes) to reuse code and extend functionality.
# 3. Polymorphism: The ability of objects of different classes to respond to the same method in different ways.
# 4. Abstraction: Simplifying complex reality by representing essential features without including unnecessary details.


# Q3. Explain why the __init__() function is used. Give a suitable example.
# The __init__() function is used as a constructor in Python classes. It is automatically called when an object is created from the class and allows us to initialize the object's attributes.
# Example:

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I'm {self.age} years old.")


person = Person("John", 25)
person.greet()


# Q4. Why self is used in OOPs?
# 'self' is used in object-oriented programming as a reference to the instance of the class. It represents the current object, allowing access to its attributes and methods within the class.
# It is the first parameter in class methods and is used to differentiate between instance variables and local variables.
# Without 'self', we wouldn't be able to access or modify object-specific attributes and methods.


# Q5. What is inheritance? Give an example for each type of inheritance.
# Inheritance is a mechanism in object-oriented programming that allows a class (derived class or subclass) to inherit the properties and methods of another class (base class or superclass).
# It promotes code reusability and helps create a hierarchy of classes.

# Types of inheritance:
# 1. Single Inheritance:
#    - A derived class inherits from a single base class.
# Example:

class Animal:
    def speak(self):
        print("The animal makes a sound.")


class Dog(Animal):
    def speak(self):
        print("The dog barks.")


dog = Dog()
dog.speak()


# 2. Multiple Inheritance:
#    - A derived class inherits from multiple base classes.
# Example:

class Flyable:
    def fly(self):
        print("The object can fly.")


class Bird(Animal, Flyable):
    pass


bird = Bird()
bird.speak()
bird.fly()


# 3. Multilevel Inheritance:
#    - A derived class inherits from another derived class.
# Example:

class Puppy(Dog):
    def play(self):
        print("The puppy plays.")


puppy = Puppy()
puppy.speak()
puppy.play()


# 4. Hierarchical Inheritance:
#    - Multiple derived classes inherit from a single base class.
# Example:

class Cat(Animal):
    def speak(self):
        print("The cat meows.")


class Lion(Animal):
    def speak(self):
        print("The lion roars.")


cat = Cat()
cat.speak()

lion = Lion()
lion.speak()
```
